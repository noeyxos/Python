```python
# 핵심기능 Gruop By 
german = german_sample[['Credit Amount','Type of apartment']]
german.head()
```

```python
german_grouped = german['Credit Amount'].groupby(german['Type of apartment'])   # 'Type of apartment'을 기준으로 Credit Amount 그룹 짓기
```

```python
german_grouped.mean()
```

```python
# 다중 항목으로 Groupby
german = german_sample[['Credit Amount','Type of apartment','Purpose']]
german_group2 = german['Credit Amount'].groupby(
    ([german['Purpose'],                                                                #첫 번째 항목 Purpose
    german['Type of apartment']])                                                #두 번째 항목 Type of apartment
)
german_group2.mean()
```

```python
german = german_sample[['Type of apartment',
                         'Sex & Marital Status',
                        'Credit Amount']]
                        
# for문을 사용해서 Groupby
for type, g in german.groupby('Type of apartment'):                          # Type of apartment 으로 그룹을 짠다. 그룹별로 데이터를 확인하고 싶어서 type, g로 type은 따로 떼고 출력한다
  print(g.head())                                                                                   # head(n = 3)으로 head의 개수도 제어 가능
  print("-------------------------------------------------------------")    # 각 그룹별로 구분하려고 절취선을 넣어준다.
```

```python
births = pd.read_csv('https://raw.githubusercontent.com/jakevdp/data-CDCbirths/master/births.csv')
births.head()
```

```python
births['decade'] = births['year']  // 10 * 10     # 새로운 컬럼 decade 추가하기   # 참고로 연산  / 는 나누기 //는 나머지는 버리는 나누기
births.head()
```

```python
births.pivot_table('births',index='decade', columns = 'gender', aggfunc='sum')   #pivot_table() : 행을 열로 바꿔주어 테이블을 만들어주는 메소드 
```

```python
births.pivot_table('births',index='decade', columns = 'gender', aggfunc='sum').plot()     #aggfunc: 집계함수. plot()  : 시각화 자동으로 x축 index , y축 columns으로 생성
```
