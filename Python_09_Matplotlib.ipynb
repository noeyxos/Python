# 한번에 두 개의 그래프를 출력하는 방법

```python
plt.figure()
```

```python
plt.subplot(2,1,1)          # rows, columns, panel number
plt.plot(x, np.sin(x))      # 그래프 속성 지정
plt.subplot(2,1,2)
plt.plot(x, np.cos(x))
```

▶ 그래프의 색상 지정법
1. color = '원하는 색상'
2. color = '단축어'
3. color = '0에서 1사이의 수'
4. color = '# RGB 색상 코드' 
5. color = '(R 농도 값,G 농도 값,B 농도 값)'
6. color = '색상이름'  - 잘안씀

```python
plt.plot(x, np.sin(x), color ='blue')                 #파란선
plt.plot(x, np.sin(x-1), color = 'k')                 #검정색의 단축어
plt.plot(x, np.sin(x-2), color = '0.75')              #농도... ?명도:  1에 가까울 수록 진한 선 0 에 가까울 수록 흰 선 
plt.plot(x, np.sin(x-3), color = '#ffdd44')           #RGB코드
plt.plot(x, np.sin(x-4), color = (1.0, 0.2, 0.3))     #RGB를 숫자로 표현  (1에 가까울 수록 진함)
plt.plot(x, np.sin(x-5), color = 'chartreuse')        #색상이름       - 잘 안씀 
```

▶ 그래프의 선 스타일 지정법

```python
# 언어로 표시
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')


# 기호로 표시 
plt.plot(x, x+5, linestyle='-')
plt.plot(x, x+6, linestyle='--')
plt.plot(x, x+7, linestyle='-.')
plt.plot(x, x+8, linestyle=':')
```

▶선 스타일과 색상 한번에 지정하기

```python
plt.plot(x,x+0, '-g')                    # solid green
plt.plot(x,x+1, '--c')                   # dashed cyan
plt.plot(x,x+2, '-.k')                   # dashed black 
plt.plot(x,x+3, ':r')                    # dotted red
```

▶세부 조정하는 법

```python
plt.plot(x, np.sin(x))      # 원래의 그래프는 x 축과 y 축의 기준범위가 다르다..(꽉채워서 그래프를 표현하기 위해 임의로 그려놓음)
plt.xlim(-1,11)             # x 축의 범위 지정 (시작, 끝)
plt.ylim(-1.5,1.5)          # y 축의 범위 지정
```

```python
plt.plot(x, np.sin(x))
plt.xlim(10,0)                   # 그래프 좌우반전
plt.ylim(1.2,-1.2)               # 그래프 상하반전
```

```python
plt.plot(x, np.sin(x))            # 한꺼번에 쓰는 방법
plt.axis([-1,11,-1.5,1.5])        # 차례대로 [x축 시작, x축 끝, y축시작, y축 끝]
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight')             # 타이트하게 출력(디폴트 값임)
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')               # x,y 비율을 같게한다. (실제 그래프의 크기를 가늠하고 싶을 때 사용)
```

#label 붙이기

```python
plt.plot(x, np.sin(x))
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
plt.plot(x, np.sin(x), '-g', label = 'sin(x)')      # label = 'sin(x)'는 그래프의 이름을 지어준 것이라 축에 바로 표기되지 않는다.
plt.plot(x, np.cos(x), ':k', label = 'cos(x)')
plt.axis('equal')
plt.legend()                                        # legend() : 각주 
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sin and Cos')
```

# 산점도의 표현 (Scatter Plot)

```python
x = np.linspace(0,10,30)
y = np.sin(x)

plt.plot(x,y, 'o' , color = 'black')    #marker  'o' 는 o 모양으로 나온다(모양바꾸기 가능.. )
```

```python
plt.style.use('bmh')                                                              #배경을 어둡게
rng = np.random.RandomState(0)                                     #seed 값: 0
for marker in ['o',',',',','x','+','v','^','<','>','s','d']:                    #'v','^','<','>' : 다양한 삼각형의 모양  
  plt.plot(rng.rand(5), rng.rand(5), marker, label='marker={0}'.format(marker)) 
  plt.legend(numpoints=1)
  plt.xlim(0,1.8)
```

```python
plt.style.use('classic')

rng = np. random.RandomState(0)                                                 # rng 는 random number generation의 약자
x = rng.randn(100)                                                                          # randn(100) :  0을 기준으로 +-100개 생성 
y = rng.randn(100)
color = rng.rand(100)                                                                      # rand(100)  (0-1)의 100개의 임의의 값을 생성
sizes = rng.rand(100) * 1000
plt.scatter(x,y, s = sizes, c=color, alpha=0.3, cmap='viridis')                 #cmap(colormap) 일관성있게 색 지정 # alpha : 투명도 0에 가까울수록 투명하다. cmap: (colormap) 지정
```

```python
from sklearn.datasets import load_iris          # 사이킷런 데이터 샘플 load_iris 패키지를 가져오기
iris = load_iris()                                              # 붓꽃 데이터
features = iris.data.T
print(iris.target_names)

plt.scatter(features[0],features[1],c=iris.target, s =features[3]*100,alpha=0.2,cmap='viridis')        #c=iris.target 품종에 따라 color가 다르게 나타남
plt.xlabel(iris.feature_names[0])         # sepal length
plt.ylabel(iris.feature_names[1])         # sepal width                         
```

# 오차 시각화 하기

```python
x = np.linspace(0,10,50)          #linspace(): 점 생성 
dy = 0.8                                   #오차범위
y = np.sin(x + np.random.randn(50))         #randn : 0을 기준으로 +- 50개 랜덤 생성

plt.errorbar(x,y, yerr=dy, fmt='.k')    #errorbar: 오차범위 출력/ yerr: y축에 대한 에러 값은 dy  / fmt(format) : .K : . 을 black으로 표시
```

```python
plt.errorbar(x,y, yerr=dy, fmt='o', color= 'black', ecolor ='lightgrey', elinewidth=3)    #  fmt='.'은 . ,fmt='o'은 원 , elinewidth : 에러바의 선 너비 조절
```
